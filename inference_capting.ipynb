# Step 1: Import Necessary Libraries
Begin by importing the necessary libraries and modules.

```python
# imports
import os
import torch
import warnings

# Suppress the specific FutureWarning
warnings.filterwarnings("ignore",
                        category=FutureWarning,
                        message=r".*clean_up_tokenization_spaces.*")

# Import the generate_caption function from get_caption.py
from get_caption import generate_caption
from src.utils import show_imagecaption

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

possible special corrections

```python
character_name = 'sbercat'
special_corrections = [
    ("s Bercat", character_name),
    ("S Bercat", character_name),
    ("cat", character_name),
    ("Cat", character_name)
]
```

#Generate Caption from Text

```python
# Parameters
input_text = "A scene where cat is presiding over a high-tech conference room. Sbercat stay at large, oval, glass table sit diverse artificial intelligence developers, each with a holographic display in front of them. The room is filled with digital screens showing complex code and AI models."

# Generate the caption
caption = generate_caption(
    input_text=input_text,
    character_name=character_name,
    special_corrections=special_corrections,
    device=device
)

print("Generated Caption:")
print(caption)
```
